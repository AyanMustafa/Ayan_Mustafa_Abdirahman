```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('customer_data_records.csv')
df.shape
```

```python
df.head(50)
```

```python
choice = df[['Annual Income (k$)','Spending Score (1-100)']]
```

```python
kmeans = KMeans(n_clusters=2, random_state=0)
kmeans.fit(choice)
```

```python
df['Cluster'] = kmeans.labels_
df.tail(10)
```

```python
plt.figure(figsize=(15,8))
sns.scatterplot(data=df, x='Annual Income (k$)', y='Spending Score (1-100)', hue='Cluster')
plt.title("Annual Income against Spending Score")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.show()
```
